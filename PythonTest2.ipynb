```python
########### The number of times bob occurs in azcbobobegghakl #############
s = "azcbobobegghakl"
substring = 'bob'
count = 0
for i in range(len(s) - len(substring) + 1):
    if s[i:i + len(substring)] == substring:
        count += 1
print("Number of times bob occurs is: ", count)
```

```python
###Write a program that prints the longest substring of s in which the letters occur in alphabetical order. For example, if s = 'azcbobobegghakl' ###

s = 'azcbobobegghakl'
current_substring = s[0]
longest_substring = s[0]
for char in s[1:]:
  if char >= current_substring[-1]:
    current_substring += char
  else:
      current_substring = char
  if len(current_substring) > len(longest_substring):
        longest_substring = current_substring
print("Longest substring in alphabetical order is:", longest_substring)
```

```python
s = 'azcbobobegghakl'
longest = 0
longestSubstring = ""
for a in range(1, len(s)):
  for char in range(len(s) + 1):
    substring = s[char: (char + a)]
    substringSort = "".join(sorted(substring))
    if substring == substringSort:
      if len(substring) > longest:
        longestSubstring = substring
        print('Longest substring in alphabetical order is: ' + str(longestSubstring))
```
